```python
import numpy as np 
import pandas as pd 
import plotly.express as px
from generators import simulate_cauchy_coloured_noise


def acf(tau: float) -> float: 
    half_life=64
    decay_constant = np.log(2) * half_life**-2
    return np.exp(-decay_constant*tau**2)


Z = simulate_cauchy_coloured_noise(
    autocorrelation_function=acf, 
    size=2**20
)

px.line(
    pd.DataFrame(Z[:1000],columns=["Simulation"]),
    title="Cauchy coloured noise", 
).show()
```
